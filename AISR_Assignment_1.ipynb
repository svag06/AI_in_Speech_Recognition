```python
from pydub.utils import mediainfo
filename = "/content/drive/MyDrive/AISR_assignments/audio_sample.mp4"
audio_sample = mediainfo(filename)
sample_rate = int(audio_sample['sample_rate'])
print(f"Sample rate of the audio file: {sample_rate} Hz")
```

```python
def get_audio_duration(filename):
    # Load the audio file
    y, sr = librosa.load(filename, sr=sample_rate)

    # Calculate the duration in seconds
    duration = librosa.get_duration(y=y, sr=sr)

    return duration

audio_duration = get_audio_duration(filename)
print(f"The duration of the audio file is {audio_duration:.2f} seconds.")
```

```python
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
#playing the original audio by changing the sample rate
def extract_and_play_segment(filename, start_time, end_time, sampling_rate):
  y, sr = librosa.load(filename, sr=sampling_rate)
  segment = y[int(start_time * sr):int(end_time * sr)]

  global reversed_segment
  reversed_segment = np.flip(segment)
  plt.figure(figsize=(15, 4))
  plt.subplot(2, 1, 1)
  librosa.display.waveshow(segment, sr=sr)
  plt.title(f'Original Audio Segment ({start_time}s - {end_time}s)')

  plt.subplot(2, 1, 2)
  librosa.display.waveshow(reversed_segment, sr=sr)
  plt.title('Reversed Audio Segment')
  plt.tight_layout()
  plt.show()

  audio_player = Audio(data=np.array(segment), rate=sr)
  return audio_player

start_time = 0
end_time = 3.75
sampling_rate = sample_rate
audio_player = extract_and_play_segment(filename, start_time, end_time, sampling_rate)
print(f'sample of original audio file, from {start_time} seconds to {end_time} seconds')
audio_player

```

```python
#playing the reverse of original audio by changing the sample rate
audio_player2 = Audio(data=np.array(reversed_segment), rate=sr)
reversed_audio_player = audio_player2
print(f'sample of  reversed original audio file, from {start_time} seconds to {end_time} seconds')
reversed_audio_player
```
