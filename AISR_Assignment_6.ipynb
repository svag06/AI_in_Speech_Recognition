```python
#vowel-1 (I)
from IPython.display import Audio
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

def analyze_audio_segment(filename, start_time, end_time):
    """
    Analyzes an audio segment from a given file, displaying its waveform and amplitude spectrum.

    Parameters:
        filename (str): Path to the audio file.
        start_time (float): Start time of the segment in seconds.
        end_time (float): End time of the segment in seconds.

    Returns:
        Audio: IPython display object for playing the audio segment.
    """
    # Load audio file and extract segment
    audio_sample = mediainfo(filename)
    sample_rate = int(audio_sample['sample_rate'])
    y, sr = librosa.load(filename, sr=sample_rate)
    segment = y[int(start_time * sr):int(end_time * sr)]

    # Display waveform
    plt.figure(figsize=(5, 4))
    plt.subplot(2, 1, 1)
    librosa.display.waveshow(segment, sr=sr)
    plt.title('Waveform')

    # Perform FFT on the segment
    fft_result = np.fft.fft(segment)

    # Plot amplitude spectrum
    plt.subplot(2, 1, 2)
    plt.plot(fft_result)
    plt.title('Amplitude Spectrum')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.grid(True)

    # Play the audio segment
    audio_player = Audio(segment, rate=sr)

    # Show the plots
    plt.tight_layout()
    plt.show()

    return audio_player

analyze_audio_segment("/content/drive/MyDrive/AISR_assignments/speech 5 word.mp4", 1.5, 1.85)
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the audio file
filename = "/content/drive/MyDrive/AISR_assignments/speech 5 word.mp4"
y, sr = librosa.load(filename, sr=sample_rate)

# Generate the spectrogram
spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=512, n_mels=128)

# Convert to decibels
spectrogram_db = librosa.power_to_db(spectrogram, ref=np.max)

# Plot the spectrogram
plt.figure(figsize=(12, 6))
librosa.display.specshow(spectrogram_db, sr=sr, hop_length=512, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()
```
